```python
synonyms_dict = {}

for _, row in synonym_df.iterrows():
    field = row["标准字段"]
    synonyms = str(row["同义词"]).split("、")
    synonyms_dict[field] = synonyms
```

```python
#同义词字典反向化
reverse_dict = {}
for filed in synonyms_dict:
    for w in synonyms_dict[filed]:
        reverse_dict[w] = filed
list(reverse_dict.items())[:5]
```
